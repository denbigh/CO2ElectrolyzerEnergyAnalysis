## Figure A: Electrolyzer Energy Requirements vs Gas Sep Energy Requirements

## Figure C. Break Even Voltage for BPM Electrolyzer
Find voltage at which extra voltage drop from BPM exactly matches downstream gas separation.

```python
N = 100
Second_Law_Efficiency_vals = np.linspace(0.02, 0.5, N)
Break_Even_Voltage_vals = np.zeros(N)



for ii in range(N):
    def Energy_Difference(ΔV_BPM):
        AEM_Energy = MassBalance(ndot_CO2R_Product_Stream_8 = ndot_C2H4_Product,          #Set Basis for rate of production of reduced product (either CO or C2H4), mol/s (Defaults to 100t C2H4/day)
                Second_Law_Efficiency_Gas_Sep = Second_Law_Efficiency_vals[ii],       #Second Law Efficiency of Gas Separation Processes. Assumed to be identical for cathode- and anode-side separation.
                R = 8.314,                                 #Ideal Gas Constant, J/K.mol
                T = 313.15,                                #Temperature, K
                p = 1e5,                                   #Gas Pressure, Pa
                Single_Pass_Conversion_CO2 = 0.125,        #Overall Single Pass Conversion of CO2
                DeltaV = 2.0,                              #Overall Voltage Drop over Electrolyzer, V
                n_CO2R = 12,                               #Number of electrons per mol of CO2R product (12 for C2H4, 2 for CO)
                s_CO2R = 2,                                #Number of CO2 molecules per mol of CO2R product (2 for C2H4, 1 for CO)
                n_H2 = 2,                                  #Number of electrons per mol of H2 produced
                MW_CO2R = 0.028,                           #Molecular Weight, CO2R product, kg/mol
                F = 96485,                                 #Faraday's Constant, C/mol
                Charge_Carbonate_Ion = 2,                  #Charge on Carbonate Ion
                n_OER = 4,                                 #Number of electrons per mol of O2 produced
                Fraction_Charge_Carried_by_CO32 = 1.0,     #Fraction of charge carried by CO32- ions. About 1 for AEM, about 0 for BPM.
                FE_H2_0 = 0.1,                             #Faradaic Efficiency of H2. if Use_Hawks_Model is false, this will just equal the FE of H2. Else, we will take this to be the FE of H2 at very low single pass conversion, when the CO2 concentration is uniform in the reactor and FE_H2 is minimized. 
                i_CO2R_0 = 4000.0,                         #CO2R Partial Current Density, A/m2. Will be used in different ways depending on status of Use_Hawks_Model.
                User_Specify_FE_iCO2R = True)['E_total']
        
        BPM_Energy = MassBalance(ndot_CO2R_Product_Stream_8 = ndot_C2H4_Product,          #Set Basis for rate of production of reduced product (either CO or C2H4), mol/s (Defaults to 100t C2H4/day)
                Second_Law_Efficiency_Gas_Sep = Second_Law_Efficiency_vals[ii],       #Second Law Efficiency of Gas Separation Processes. Assumed to be identical for cathode- and anode-side separation.
                R = 8.314,                                 #Ideal Gas Constant, J/K.mol
                T = 313.15,                                #Temperature, K
                p = 1e5,                                   #Gas Pressure, Pa
                Single_Pass_Conversion_CO2 = 0.5,          #Overall Single Pass Conversion of CO2
                DeltaV = 2.0 + ΔV_BPM,                     #Overall Voltage Drop over Electrolyzer, V
                n_CO2R = 12,                               #Number of electrons per mol of CO2R product (12 for C2H4, 2 for CO)
                s_CO2R = 2,                                #Number of CO2 molecules per mol of CO2R product (2 for C2H4, 1 for CO)
                n_H2 = 2,                                  #Number of electrons per mol of H2 produced
                MW_CO2R = 0.028,                           #Molecular Weight, CO2R product, kg/mol
                F = 96485,                                 #Faraday's Constant, C/mol
                Charge_Carbonate_Ion = 2,                  #Charge on Carbonate Ion
                n_OER = 4,                                 #Number of electrons per mol of O2 produced
                Fraction_Charge_Carried_by_CO32 = 1e-3,    #Fraction of charge carried by CO32- ions. About 1 for AEM, about 0 for BPM.
                FE_H2_0 = 0.1,                             #Faradaic Efficiency of H2. if Use_Hawks_Model is false, this will just equal the FE of H2. Else, we will take this to be the FE of H2 at very low single pass conversion, when the CO2 concentration is uniform in the reactor and FE_H2 is minimized. 
                i_CO2R_0 = 4000.0,                         #CO2R Partial Current Density, A/m2. Will be used in different ways depending on status of Use_Hawks_Model.
                User_Specify_FE_iCO2R = True)['E_total']   #If true, use fixed values for electrolyzer performance as specified by user. If false, use Steve Hawk's modela to calcualate how electrolyzer performance changes with single pass conversion. 
        
        return AEM_Energy - BPM_Energy
    
    sol = optimize.root_scalar(Energy_Difference, bracket=[0.0, 3.0], method='brentq')
    
    Break_Even_Voltage_vals[ii] = sol.root
    
plt.figure(dpi=100)
#plt.figure(dpi=300)
plt.plot(Second_Law_Efficiency_vals, Break_Even_Voltage_vals, color = 'k')
plt.xlim([0.0, 0.5])
plt.ylim([0.0,0.5])
plt.xticks(size = 13)
plt.yticks(size = 13)
plt.fill_between(Second_Law_Efficiency_vals, 0, Break_Even_Voltage_vals, color = ColorSchemePastel2[1])
plt.fill_between(np.linspace(0,0.02,2), 0, 1, color = ColorSchemePastel2[1])
plt.fill_between(Second_Law_Efficiency_vals, Break_Even_Voltage_vals, 1, color =  ColorSchemePastel2[4])
plt.text(0.08, 0.07, "BPM \n Preferred", ha="center", va="center", size=14)
plt.text(0.29, 0.25, "Gas Separation \n Preferred", ha="center", va="center", size=14)
plt.ylabel("Break Even Voltage (V)", size = 13)
plt.xlabel("Gas Separation Second Law Efficiency", size = 13)
plt.gca().set_aspect(1.0)
#plt.savefig("Figure3b.pdf")
 
```

## Figure H. Optimal Single Pass Conversion

Here we create histograms for optimal single pass conversion, as in Figure 3 in the manuscript. We include this alternative as a more accessible demonstration of the procedure, and also because Figures S2-S4 of the SI were created in this manner (Figures I-K below). We set NumSimulations = 1000 to minimize computation time; in the final plots, NumSimulations = 100,000.

Note that if the single pass consumption is ever larger than 1 (this occurs when the single pass conversion is too large or the FE_C2H4 is too small) the model returns an error associated with taking a negative of a logarithm (see Eq. S3 in the SI). This error is ocassionally encountered when the model is scanning the parameter space for the optimal single pass conversion, but it does not affect identification of of the correct and physically-meaningful solution.
